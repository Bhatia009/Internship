WOMEN CLOTH REVIEWS PREDICTION WITH MULTI NOMINAL NAIVE BAYES

 Objective: To understand the classification with discrete features(such as word counts for text classification).

The multinominal naive bayes classifier is suitable for such classification.It normally requires integer feature counts. However, in general, fractional counts such as tf-idf may also work.

DATA SOURCE:
 Ybi Foundation

IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

IMPORT DATASET

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/ProjectHub-MachineLearning/main/Women%20Clothing%20E-Commerce%20Review.csv')
```

```python
df.head()
```

DESCRIBE DATA

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

DATA VISUALIZATION

DATA PREPROCESSING

MISSING VALUES

Remove missing values in Review Columns with no Review Text

```python
df.isna().sum()
```

```python
df[df['Review']==""]=np.NaN
df['Review'].fillna("No Review",inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

DEFINE TARGET(y) AND FEATURE (X)

```python
df.columns
```

```python
x = df['Review']
y = df['Rating']
```

```python
df['Rating'].value_counts
```

TRAIN_TEST_SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.7, stratify = y, random_state = 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

FEATURE TEXT CONVERSION TO TOKENS

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase = True, analyzer='word', ngram_range=(2,3), stop_words = 'english',max_features=5000)
```

```python
x_train = cv.fit_transform(x_train)
```

```python
cv.get_feature_names_out()
```

```python
x_train.toarray()
```

```python
x_test = cv.fit_transform(x_test)
```

```python
cv.get_feature_names_out()
```

```python
x_test.toarray()
```

MODEL TRAINING

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(x_train,y_train)
```

MODEL PREDICTION

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

PROBABILITY OF EACH PREDICTED CLASS

```python
model.predict_proba(x_test)
```

MODEL EVALUATION

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

RECATEGORIES RATINGS AS POOR(0) AND GOOD(1)

```python
df['Rating'].value_counts()
```

Re-rating 1,2,3 as 0 and 4,5 as 1

```python
df.replace({'Rating':{1:0, 2:0, 3:0, 4:1, 5:1}}, inplace = True)
y = df['Rating']
x = df['Review']
```

TRAIN_TEST_SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.7, stratify = y, random_state = 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

FEATURE TEXT CONVERSION TO TOKENS

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase = True, analyzer='word', ngram_range=(2,3), stop_words = 'english',max_features=5000)
```

```python
x_train = cv.fit_transform(x_train)
```

```python
x_test = cv.fit_transform(x_test)
```

MODEL RE-TRAIN

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB
```

```python
model.fit(x_train,y_train)
```

MODEL PREDICTION

```python
y_pred = model.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

MODEL EVALUATION

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

EXPLANATION:

Multinominal Naive bayes algorithm is a probabilistic learning method mostly used in Natural Language Processing. Based on the bayes theorem it enables prediction of tag of text, assuming features to be conditionally independent calculating the probability.

Such as in this women cloth Review prediction we founded the ratings and reviews for the clothes are good or bad by all the steps of a machine learning model.
